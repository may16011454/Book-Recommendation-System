 - shows there are only 899 users who have rated more than 200 books

```python
books.head()
```

```python
ratings_with_books = ratings.merge(books, on = "ISBN")
```

```python
ratings_with_books.head()
```

- shows only the books that have been rated more than 200 times

```python
ratings_with_books.shape
```

```python
num_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
num_rating.head()
```

- shows how many ratings each book has got so we can take the books that have less than 50 ratings out of the dataframe

```python
num_rating.rename(columns={"rating": "num_of_rating"}, inplace=True)
```

```python
num_rating.head()
```

```python
final_rating = ratings_with_books.merge(num_rating, on= 'title')
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['num_of_rating']>=50]
```

```python
final_rating.sample(10)
```

- shows all of the books that have more than 50 ratings

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'], inplace=True)
```

```python
final_rating.shape
```

```python
final_rating
```

<h3> Pivot Table </h3>

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot
```

- shows each user and what they have rated for each book

```python
book_pivot.shape
```

- this shows that there are 742 individual books and 888 users who have rated books

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors = 6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

- this shows the recommendations for the book that is on row 237 

```python
book_pivot.index[3]
```

```python
book_pivot.index
```

```python
book_name = book_pivot.index
```

```python
import pickle
pickle.dump(model, open('artifacts/model.pkl', 'wb'))
pickle.dump(book_name, open('artifacts/book_name.pkl', 'wb'))
pickle.dump(final_rating, open('artifacts/final_rating.pkl', 'wb'))
pickle.dump(book_pivot, open('artifacts/book_pivot.pkl', 'wb'))
```

- saves the model to a local file

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors = 6)
    
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```

```python
book_name = "Winter Solstice"
recommend_book(book_name)
```
